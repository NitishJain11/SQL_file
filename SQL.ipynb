```python
# Q1->What is a database ? Differentiate between SQL and NoSQL database?

# Answer->

# A database is an organized collection of data that is stored and managed in a way that allows for efficient retrieval,
# manipulation, and querying of data. Databases are widely used in various applications such as e-commerce, finance, 
# healthcare, and social media to store and manage large amounts of data.

# SQL and NoSQL are two different types of database management systems with different approaches to organizing and retrieving data.

# SQL (Structured Query Language) databases are relational databases that use a tabular schema to organize data into rows and columns.
# SQL databases enforce strict data consistency and integrity rules, which ensure that data is stored and retrieved accurately. 
# SQL databases use SQL as the query language to retrieve and manipulate data.
# SQL databases are widely used in traditional enterprise applications and data warehousing.

# NoSQL (Not Only SQL) databases, on the other hand, use a non-relational approach to data management.
# NoSQL databases can handle unstructured, semi-structured, and structured data and are designed to scale horizontally,
#       meaning they can add more nodes to the database cluster as needed. 
# NoSQL databases can be classified into different categories such as document-oriented, key-value, column-family, and graph databases. 
# NoSQL databases use various query languages such as JSON, BSON, and MapReduce to retrieve and manipulate data.

# Some of the main differences between SQL and NoSQL databases are:

# Schema: SQL databases have a predefined schema that must be followed while NoSQL databases have a flexible schema.
# Scalability: NoSQL databases are designed to scale horizontally by adding more nodes to the cluster while SQL databases typically scale vertically by adding more resources to a single machine.
# Query Language: SQL databases use SQL as the query language, while NoSQL databases use various query languages such as JSON, BSON, and MapReduce.
# Data Consistency: SQL databases enforce strict data consistency and integrity rules while NoSQL databases may sacrifice consistency for better scalability and performance.
# In summary, SQL databases are best suited for applications with a structured schema and data consistency requirements, while NoSQL databases are better suited for applications with
# a flexible schema and high scalability requirements.
```

```python
# Q2->what is DDL? explain why CREATE , DROP , ALTER and TRUNCATE and used with an example.

# Answer->

# DDL stands for Data Definition Language, which is a subset of SQL (Structured Query Language) used to define the structure of a database.
# DDL statements are used to create, modify, and delete database objects such as tables, indexes, and constraints.

# 1)-CREATE Statement:
# The CREATE statement is used to create a new database object such as a table, view, or index.
# For example, the following statement creates a new table named "Customers" with columns "id", "name", "email", and "phone":

# syntax:
    
#     CREATE TABLE Customers (
#     id INT PRIMARY KEY,
#     name VARCHAR(50),
#     email VARCHAR(50),
#     phone VARCHAR(20)
# );


# 2)->DROP Statement:
# The DROP statement is used to delete a database object such as a table, view, or index. 
# For example, the following statement deletes the "Customers" table:
# syntax

# DROP TABLE Customers;

# 3)->ALTER Statement:
# The ALTER statement is used to modify the structure of an existing database object such as a table, view, or index. 
# For example, the following statement adds a new column "address" to the "Customers" table:
# syntax:
    
#     ALTER TABLE Customers
# ADD address VARCHAR(100);

# 4)->TRUNCATE Statement:
# The TRUNCATE statement is used to delete all the data from a table. It is similar to the DELETE statement, 
# but it is faster because it does not log individual row deletions.
# For example, the following statement deletes all the data from the "Customers" table:

# syntax:
#     TRUNCATE TABLE Customers;

```

```python
# Q5->Explain Primary Key and Foreign Key .

# Answer->

# In relational databases, a primary key and a foreign key are two important concepts used to establish relationships between tables.

# 1)->Primary Key:
    
# A primary key is a column or a combination of columns in a table that uniquely identifies each row in the table. 
# It is a constraint that enforces the integrity of the data and ensures that each row in the table is unique.
# Primary keys are used to establish relationships between tables and are often used as the basis for indexing data in a database.

# 2)->Foreign Key:
    
# A foreign key is a column or a combination of columns in a table that refers to the primary key of another table.
# It is used to establish relationships between tables and to ensure data integrity across related tables. 
# A foreign key constraint ensures that the values in the foreign key column(s) of one table match the values in
# the primary key column(s) of another table.
```

```python
# Q7->Give the order of execution of SQL clauses in an SQL query?

# Answer->

# In SQL, the order of execution of clauses in an SQL query is as follows:

# FROM clause: The FROM clause specifies the table(s) from which data is to be retrieved.

# JOIN clause: If the query involves a join operation, the JOIN clause is executed next. The JOIN clause specifies how two or more tables are to be joined together.

# WHERE clause: The WHERE clause is used to filter the results of a query based on specified conditions.

# GROUP BY clause: The GROUP BY clause is used to group the results of a query by one or more columns.

# HAVING clause: The HAVING clause is used to filter the results of a query based on specified conditions after the GROUP BY clause has been executed.

# SELECT clause: The SELECT clause is used to select the columns that are to be included in the results of the query.

# DISTINCT clause: The DISTINCT clause is used to eliminate duplicate rows from the results of a query.

# ORDER BY clause: The ORDER BY clause is used to sort the results of a query based on one or more columns.

# LIMIT/OFFSET clause: The LIMIT/OFFSET clause is used to limit the number of rows returned by a query and to specify the starting row for the result set.

# It is important to note that not all clauses are required in a SQL query, and some clauses may not be executed if they are not used in the query. For example,
# if a query does not include a GROUP BY clause, the HAVING clause will not be executed.
```
